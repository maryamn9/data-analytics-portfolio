```python
#Day 10 | 15 May 2026 | Maryam Naveen
#Topic: Exploratory Data Analysis (EDA)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Setting style for all charts
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

#Loading clean data
df = pd.read_csv(r'C:\Users\marya\OneDrive\Documents\GitHub\data-analytics-portfolio\day-07-sql-project\UK_Accident.csv')
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

print(f"Dataset loaded: {df.shape}")
print("Ready for EDA!")
```

```python
#Chart 1: Accidents per year — is UK getting safer?
yearly = df.groupby('Year')['Accident_Index'].count().reset_index()
yearly.columns = ['Year', 'Total_Accidents']

plt.figure(figsize=(12,6))
plt.plot(yearly['Year'], yearly['Total_Accidents'], marker='o', color='#E74C3C', linewidth=2.5)
plt.fill_between(yearly['Year'], yearly['Total_Accidents'], alpha=0.2, color='#E74C3C')
plt.title('UK Road Accidents Per Year (2005-2014)', fontsize=16, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Accidents')
plt.xticks(yearly['Year'])
plt.tight_layout()
plt.savefig('chart1_accidents_per_year.png', dpi=150)
plt.show()
print("Chart 1 saved!")
```

```python
#Chart 2: Accidents by day of week
days = {1:'Sun', 2:'Mon', 3:'Tue', 4:'Wed', 5:'Thu', 6:'Fri', 7:'Sat'}
day_counts = df['Day_of_Week'].value_counts().sort_index()
day_counts.index = [days[i] for i in day_counts.index]

plt.figure(figsize=(10,6))
bars = plt.bar(day_counts.index, day_counts.values, color=['#3498DB']*5 + ['#E74C3C'] + ['#3498DB'])
plt.title('UK Road Accidents by Day of Week', fontsize=16, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Total Accidents')
plt.tight_layout()
plt.savefig('chart2_accidents_by_day.png', dpi=150)
plt.show()
print("Chart 2 saved!")
```

```python
#Chart 3: Accident severity pie chart
severity_labels = {1: 'Fatal', 2: 'Serious', 3: 'Slight'}
severity_counts = df['Accident_Severity'].value_counts().sort_index()
severity_counts.index = [severity_labels[i] for i in severity_counts.index]

plt.figure(figsize=(8,8))
plt.pie(severity_counts.values, labels=severity_counts.index,
        autopct='%1.1f%%', colors=['#E74C3C','#F39C12', '#2ECC71'],
        startangle=90)
plt.title('UK Road Accident Severity', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('chart3_severity_pie.png', dpi=150)
plt.show()
print("Chart 3 saved!")
```

```python
#Chart 4: Urban vs Rural
urban_rural = {1: 'Urban', 2: 'Rural', 3: 'Unclassified'}
ur_counts = df['Urban_or_Rural_Area'].value_counts().sort_index()
ur_counts.index = [urban_rural[i] for i in ur_counts.index]

plt.figure(figsize=(8,6))
plt.bar(ur_counts.index, ur_counts.values, color=['#3498DB', '#27AE60', '#95A5A6'])
plt.title('UK Road Accidents — Urban vs Rural', fontsize=16, fontweight='bold')
plt.xlabel('Area Type')
plt.ylabel('Total Accidents')
plt.tight_layout()
plt.savefig('chart4_urban_rural.png', dpi=150)
plt.show()
print("Chart 4 saved!")
```

```python
#Chart 5: Weather conditions (top 5)
weather = df['Weather_Conditions'].value_counts().head(5)

plt.figure(figsize=(12,6))
plt.barh(weather.index.astype(str), weather.values, color='#9B59B6')
plt.title('Top 5 Weather Conditions During Accidents', fontsize=16, fontweight='bold')
plt.xlabel('Total Accidents')
plt.tight_layout()
plt.savefig('chart5_weather.png', dpi=150)
plt.show()
print("Chart 5 saved!")
```
