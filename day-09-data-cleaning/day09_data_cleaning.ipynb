```python
#Day 09 | 14 May 2026 | Maryam Naveen
#Topic: Data Cleaning — nulls, dtypes, duplicates

import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\marya\OneDrive\Documents\GitHub\data-analytics-portfolio\day-07-sql-project\UK_Accident.csv')

print(f"Shape: {df.shape}")
print("Dataset loaded!")
```

```python
#Step 1: Checking missing values as percentages
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
print(missing_df[missing_df['missing_count'] > 0])
```

```python
#Step 2: Dropping columns with too many missing values (over 30%)
df_clean = df.drop(columns=['Junction_Control', 'LSOA_of_Accident_Location'])
print(f"Columns before: {df.shape[1]}")
print(f"Columns after: {df_clean.shape[1]}")
```

```python
#Step 4: Checking for duplicates
print(f"Duplicate rows: {df_clean.duplicated().sum()}")
```

```python
#Step 5: Fixing Date column to datetime
df_clean['Date'] = pd.to_datetime(df_clean['Date'], dayfirst=True)
print(df_clean['Date'].dtype)
print(df_clean['Date'].head())
```

```python
#Step 6: Final clean dataset summary
print(f"Original shape: {df.shape}")
print(f"Clean shape: {df_clean.shape}")
print(f"Missing values remaining: {df_clean.isnull().sum().sum()}")
print(f"Date range: {df_clean['Date'].min()} to {df_clean['Date'].max()}")
```

```python
#Checking what's still missing
remaining = df_clean.isnull().sum()
print(remaining[remaining > 0].sort_values(ascending=False))
```

```python
#Dropping remaining high-missing columns
cols_to_drop = ['Carriageway_Hazards', 'Special_Conditions_at_Site']
cols_to_drop = [c for c in cols_to_drop if c in df_clean.columns]
df_clean = df_clean.drop(columns=cols_to_drop)

#Filling remaining small missing values
df_clean = df_clean.copy()
df_clean['Time'] = df_clean['Time'].fillna('Unknown')
df_clean['Location_Easting_OSGR'] = df_clean['Location_Easting_OSGR'].fillna(df_clean['Location_Easting_OSGR'].median())
df_clean['Longitude'] = df_clean['Longitude'].fillna(df_clean['Longitude'].median())
df_clean['Pedestrian_Crossing-Physical_Facilities'] = df_clean['Pedestrian_Crossing-Physical_Facilities'].fillna(0)
df_clean['Pedestrian_Crossing-Human_Control'] = df_clean['Pedestrian_Crossing-Human_Control'].fillna(0)

print(f"Final shape: {df_clean.shape}")
print(f"Missing values remaining: {df_clean.isnull().sum().sum()}")
print("Data is clean!")
```

```python
#Summary
print("=" * 50)
print("DATA CLEANING SUMMARY")
print("=" * 50)
print(f"Original: {df.shape[0]:,} rows, {df.shape[1]} columns")
print(f"Clean:    {df_clean.shape[0]:,} rows, {df_clean.shape[1]} columns")
print(f"Columns dropped: 4 (over 97% missing)")
print(f"Missing values: {df_clean.isnull().sum().sum()}")
print(f"Duplicates: {df_clean.duplicated().sum()}")
print(f"Date type: {df_clean['Date'].dtype}")
```
